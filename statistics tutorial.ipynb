```python
desc=df.describe().transpose()
desc
```

```python
plt.boxplot(df.radius_mean)
```

```python
desc['iqr']=desc['75%']-desc['25%']
desc['uppwhisk']=desc['75%']+(1.5*(desc['iqr']))
desc['lowwhisk']=desc['25%']-(1.5*(desc['iqr']))

```

```python
import thinkplot
import thinkstats2
```

```python
df_be=df[df.diagnosis=="M"]
cdf=thinkstats2.Cdf(df_be.radius_mean)
thinkplot.Cdf(cdf)
```

```python
print("Outliers: ",df_be[(df_be.radius_mean < desc['lowwhisk'].radius_mean) | (df_be.radius_mean > desc['lowwhisk'].radius_mean )].radius_mean.values)
```

```python
df.drop(df[(df['radius_mean']<desc.loc['radius_mean'].lowwhisk)|
        (df['radius_mean']>desc.loc['radius_mean'].uppwhisk)].index,inplace=True)
```

```python
plt.boxplot(df.radius_mean)
```
